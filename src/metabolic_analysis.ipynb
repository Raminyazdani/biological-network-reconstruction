# Metabolic Network Analysis
## Multi-Model Processing - Path Fixed

```python
import cobra
import pandas as pd
import numpy as np
import glob
```

```python
# Load multiple models - FIXED PATH
model_files = glob.glob("models/*.xml")  # Corrected to work from src/ directory
print(f"Found {len(model_files)} models")
```

```python
# Process each model
for model_file in model_files:
    model = cobra.io.read_sbml_model(model_file)
    solution = model.optimize()
    print(f"{model.id}: {solution.objective_value}")
```
